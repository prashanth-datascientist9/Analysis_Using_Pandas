```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
#asc sort - not in place
movies.title.sort_values()
```

```python
#desc sort - not in place
movies.title.sort_values(ascending=False)
```

```python
#Asc sort - not in place - sorts the entire dataframe
movies.sort_values('title')
```

```python
#desc sort - not in place - sorts entire dataframe
movies.sort_values('duration', ascending=False)
```

```python
movies.head()
```

```python
#sort by multiple columns - sorts content rating and then within that by duration
movies.sort_values(['content_rating', 'duration'], ascending=[True, False])
```
