```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
#get all columns of row 0
ufo.loc[0, :]
```

```python
#get all columns from rows 0, 1, 2
ufo.loc[[0,1,2], :]
```

```python
#get all columns from rows 0, 1, 2 - loc is inclusive on both sides
ufo.loc[0:2, :]
```

```python
#column selection
ufo.loc[:, 'City']
```

```python
#multiple columns selection
ufo.loc[0:2, 'City':'State']
```

```python
#filter rows by a value
ufo.loc[ufo.City=='Oakland', :]
```

```python
#get rows by position - inclusive of first number and exclusive of second number
ufo.iloc[:, 0:4]
```

```python
#get rows by position - inclusive of start index and exclusive of end index
ufo.iloc[0:3, :]
```
